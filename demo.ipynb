```python
# Gradient Descent to find local minima of y = (x + 3)^2

# Step 1: Initialize parameters
current_x = 2           # starting point
rate = 0.1              # learning rate
precision = 1e-9        # stop condition
delta_x = 1
max_iterations = 1000
iteration = 0

# Step 2: Define function and its derivative
def slope(x):
    return 2 * (x + 3)   # derivative

def value_y(x):
    return (x + 3)**2    # function

# Step 3: Gradient Descent Loop
while delta_x > precision and iteration < max_iterations:
    prev_x = current_x
    current_x = prev_x - rate * slope(prev_x)
    delta_x = abs(current_x - prev_x)
    iteration += 1

print(f"Local minima occurs at x = {current_x:.6f}")
print(f"Value of y = {value_y(current_x):.6f}")
print(f"Total iterations = {iteration}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

x_curve = np.linspace(-6, 4, 200)
y_curve = (x_curve + 3)**2

plt.plot(x_curve, y_curve, label='y=(x+3)^2', color='blue')
plt.scatter(-3, 0, color='green', s=100, label='Local Minima')
plt.title("Gradient Descent - Finding Local Minima")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
